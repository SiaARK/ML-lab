```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
n_points = 100
n_dim = 3
data = np.random.rand(n_points,n_dim)
print(data)
```

```python
# scatter plot

plt.figure(figsize=(8,6))
plt.scatter(data[:,0],data[:,1],c='r',marker='o',label="dimension 1 v/s dimension 2")
plt.xlabel("dimension 1")
plt.ylabel("dimension 2")
plt.title("Scatter plot for 2D data")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# box plot

plt.figure(figsize=(8,6))
sns.boxplot(data=data)
plt.xlabel("dimensions")
plt.ylabel("values")
plt.title("Box plot of n-dimensional data")
plt.grid(True)
plt.show()
```

```python
# heat map

corr_matrix = np.corrcoef(data,rowvar = False)
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,annot=True,cmap = 'coolwarm',fmt = ".2f",xticklabels=range(n_dim),yticklabels=range(n_dim))
plt.title("Heatmap of correlation between dimensions")
plt.xlabel("dimensions")
plt.ylabel("dimensions")
plt.show()
```

```python
# contour plot

if n_dim >= 3:
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111,projection="3d")
    [X, Y] = np.meshgrid(data[:,0], data[0:,1]) 
    Z = np.sin(Y / 4) * np.cos(X / 2)
    ax.contour(data[:,0],data[:,1],Z,100,cmap='viridis')
    ax.set_xlabel("dimension 1")
    ax.set_ylabel("dimension 2")
    ax.set_zlabel("dimension 3")
    plt.title("contour plot of 3D data")
    plt.show()
```

```python
# 3D surface plot

if n_dim >= 3:
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111,projection="3d")
    ax.plot_trisurf(data[:,0],data[:,1],data[0:,2],cmap='viridis',edgecolor='black')
    ax.set_xlabel("dimension 1")
    ax.set_ylabel("dimension 2")
    ax.set_zlabel("dimension 3")
    plt.title("3D surface plot of 3D data")
    plt.show()
```
